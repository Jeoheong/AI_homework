```python
# P2
# use DFS algorithm to search the mango buyer
class Graph():
    def __init__(self,nodes,sides):
        self.adjacent  = {}
        self.side = []
        for node in nodes:
            for side in sides:
                u,v = side
                if node == u: self.side.append(v)
                elif node == v: self.side.append(u)
            self.adjacent[node] = self.side
            self.side = []

# DFS algorithm
    def DFS(self,node0):
        stack, visited = [], []
        stack.append(node0)
        while stack:
            node = stack.pop()
            if node[-1] == "m":
                print ("We have find the mango buyer! His name is " + node)
                break
            visited.append(node)
            for adjacent in self.adjacent[node]:
                if adjacent not in stack and adjacent not in visited:
                    stack.append(adjacent)
        return visited

# main function
if __name__ == "__main__":
    nodes = ['you','claire','thom','jonny','alice','bob','peggy','anuj']
    sides = [
        ('you','claire'),('you','bob'),('you','alice'),('claire','thom'),('claire','jonny'),('bob','peggy'),('bob','anuj'),('alice','peggy')
    ]
    G = Graph(nodes, sides)
    G.DFS('you')
```

```python
# P5
import math
goal = 85
state = [-1]*(goal + 1)
state[2] = 0
for k in range(2, int(goal/2)+1):
    if state[k] < 0: continue
    for pos, cost in [(k*2, k), (k*2+1, math.floor(k/2)), (k*2+2, k+2)]:
        if pos > goal: continue
        if state[pos] == -1 or state[pos] > state[k] + cost:
            state[pos] = state[k] + cost
#    print (state)
print (state[goal])
```
